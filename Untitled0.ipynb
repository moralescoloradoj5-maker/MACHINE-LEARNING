```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,2*np.pi,100)

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.plot(x,np.sin(x))
plt.title('Funcion Seno')

plt.subplot(1,2,2)
plt.plot(x,np.cos(x))
plt.title('Funcion Coseno')

plt.show()
```

```python
x = np.linspace(-10,10,100)
y = x**2

plt.figure()
plt.plot(x,y, label='y=x^2')
plt.title("Funcion cuadratica")
plt.xlabel("x")
plt.ylabel("y")

plt.grid(True)
plt.legend()
plt.show()
```

```python
X = np.linspace(0,5,50)
Y = np.exp(X)
plt.figure()
plt.plot(X,Y)

plt.title("Crecimiento Exponencial")
plt.xlabel("x")
plt.ylabel("e^x")

plt.savefig("grafica_exponencial.png", dpi=300)
plt.show()
```

```python
def relu(x):
  return np.maximum(0,x)

x=np.array([1.0, -1.0])

for i in range(3):
  x=relu(np.dot(np.eye(2),x))
  print(f'Capa(i+1):{x}')
```

```python
import numpy as np

x= np.array([1.0, 2.0, -1.0])
w= np.array([0.5, -0.3, 0.8])
b= 0.1

z= np.dot(w, x) + b
print("Salida Lineal", z)
```

```python
def relu(z):
    return np.maximum(0, z)

a = relu(z)
print("Salida Relu", a)
```
